```python
import pandas as pd
import numpy as np

from google.colab import files
```

```python
cat = pd.read_csv(
    "catboost_prob.csv"
)

xgb = pd.read_csv(
    "xgb_prob.csv"
)

sample = pd.read_csv(
    "sample_submission.csv"
)

print(cat.shape)
print(xgb.shape)
```

```python
assert (
    cat["id"] == xgb["id"]
).all()

print(
    "IDs match"
)
```

```python
class_cols = [
    col
    for col in cat.columns
    if col != "id"
]

class_cols
```

```python
blend_50 = cat.copy()

blend_50[class_cols] = (
    0.50 * cat[class_cols]
    +
    0.50 * xgb[class_cols]
)
```

```python
blend_60 = cat.copy()

blend_60[class_cols] = (
    0.40 * cat[class_cols]
    +
    0.60 * xgb[class_cols]
)
```

```python
blend_70 = cat.copy()

blend_70[class_cols] = (
    0.30 * cat[class_cols]
    +
    0.70 * xgb[class_cols]
)
```

```python
def create_submission(
    prob_df,
    filename
):

    preds = prob_df[class_cols] \
        .idxmax(axis=1)

    submission = pd.DataFrame({

        "id":
            prob_df["id"],

        "class":
            preds

    })

    submission.to_csv(
        filename,
        index=False
    )

    return submission
```

```python
sub_50 = create_submission(
    blend_50,
    "submission_ensemble_50_50.csv"
)

sub_60 = create_submission(
    blend_60,
    "submission_ensemble_40_60.csv"
)

sub_70 = create_submission(
    blend_70,
    "submission_ensemble_30_70.csv"
)
```
